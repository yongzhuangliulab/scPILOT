```python
import anndata as ad
adata = ad.read_h5ad('IFNGR2.h5ad')
adata.obs
```

```python
import scanpy as sc
sc.pp.pca(adata)
sc.pl.pca_variance_ratio(adata, log = True, n_pcs = 50)
```

```python
sc.pp.neighbors(adata, n_pcs = 20)
sc.tl.umap(adata)
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
sns.set_theme(style = 'white', font = 'Arial', font_scale = 2)
adata.obs = adata.obs.replace('non-targeting', 'Non-targeting')
cell_labels = sorted(adata.obs['cell_type'].unique().tolist())
adata.obs.rename(columns = {'target_gene': 'Target_gene'}, inplace = True)
adata.obs.rename(columns = {'cell_type': 'Cell_type'}, inplace = True)
sc.pl.umap(adata, use_raw = False, color = 'Target_gene', palette = {
    'Non-targeting': sns.color_palette()[0],
    'IFNGR2': sns.color_palette()[1],
}, frameon = False)
plt.savefig(f'./umap_cond.jpg', dpi = 300, bbox_inches = 'tight')
plt.close()
sc.pl.umap(adata, use_raw = False, color = 'Cell_type', palette = sns.color_palette(), frameon = False)
plt.savefig(f'./umap_cell_type.jpg', dpi = 300, bbox_inches = 'tight')
plt.close()
```
